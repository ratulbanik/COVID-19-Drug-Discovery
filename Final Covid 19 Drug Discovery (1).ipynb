Check missing value percentage: a) Drop values if missing > 40-50% (or > 60-70% in some cases). b) For missing < 40%, proceed to step 2.

For categorical variables: Replace missing values with column mode.

```python
def numerical_analysis(data, x):
    """
    Function to plot a box plot for numerical variables and impute missing values.

    Parameters:
    - data: DataFrame containing the data.
    - x: Name of the numerical variable for analysis.
    """
    sns.boxplot(x=data[x])
    
    null_val = data[x].isnull().sum()
    mean_val = data[x].mean()
    median_val = data[x].median()
    
    print("mean is:", mean_val)
    print("median is:", median_val)
    
    if mean_val == 0:
        rep_val = median_val
        print("Mean value is zero. Using median to replace null values.")
    else:
        check_val = math.fmod(median_val, mean_val)
        if check_val < 0.5:
            rep_val = mean_val
            print("The difference between mean and median is not much, so mean can be used to replace null values.")
        else:
            rep_val = median_val
            print("The difference between mean and median is too much, so mean cannot be used to replace null values, thus we use the median.")
    
    print(f"Thus, {x} is a numerical variable having {null_val} null values; therefore, we replace them with {rep_val}, which is the {'mean' if rep_val == mean_val else 'median'}.")
    data[x].fillna(rep_val, inplace=True)
    print("Recheck for null values:", data[x].isnull().sum())
```

Here we will use logistic regression or simmilar techniques so we will handel the outliers ie we will drop them


```python
def handle_outliers_iqr(data):
    Q1 = np.percentile(data, 25)  # 1st quartile (25th percentile)
    Q3 = np.percentile(data, 75)  # 3rd quartile (75th percentile)
    IQR = Q3 - Q1  # Interquartile range
    
    lower_bound = Q1 - 1.5 * IQR  # Lower bound for outliers
    upper_bound = Q3 + 1.5 * IQR  # Upper bound for outliers
    
    # Replace outliers with the nearest valid value within the bounds
    data_without_outliers = np.clip(data, lower_bound, upper_bound)
    
    return data_without_outliers
```

```python
def num_cont_UVA(data, x, bin_no=10, my_color='blue', my_list=None):
    """
    A function for univariate analysis of continuous numerical variables.
    """
    if my_list is None:
        my_list = [x, "Frequency", "Histogram for " + x]

    # Calculate histogram counts and bin edges
    hist_counts, bin_edges = np.histogram(x, bins=bin_no, range=(x.min(), x.max()), density=False)

    # Find the bin with maximum frequency
    max_freq_bin = np.argmax(hist_counts)

    # Plot the histogram
    x_bins = pd.cut(np.array(x), bins=bin_no, right=False)
    x_bins = x_bins.astype('str')
    x_plot = plt.hist(x_bins, color=my_color)
    plt.xticks(rotation=90, ha='right')
    plt.xlim(0, bin_no)
    plt.title(my_list[2])
    plt.ylabel(my_list[1])
    plt.xlabel(my_list[0])
    plt.show()

    # Display the bin with maximum frequency separately
    print(f"Bin with maximum frequency: {bin_edges[max_freq_bin]} - {bin_edges[max_freq_bin + 1]}")
```

```python
# Assuming 'covid' is your DataFrame containing the data
for col in columns:
    num_cont_UVA(data=covid, x=covid[col], bin_no=10, my_list=[col, "Frequency", "Histogram for " + col])
```
